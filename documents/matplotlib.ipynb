# Matplotlib

```python
import numpy as np
from sklearn import datasets

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 

### Load Dataset
iris = datasets.load_iris()

X = iris.data
y = iris.target

#print(X)
```

```python
pca = PCA(n_components=2)
pca.fit(X)
principalComponents = pca.fit_transform(X)
```

## Scatterplot

We use the target `y` to color each dot.

```python
plt.scatter(principalComponents[:, 0], principalComponents[:, 1],c=y)
plt.show()
```

Scatterplot with annotations (some of them)

```python
plt.scatter(principalComponents[:, 0], principalComponents[:, 1],c=y)
for idx in range(80,100):
    plt.annotate(str(idx), xy=(principalComponents[idx, 0], principalComponents[idx, 1]))
plt.show()
```

## Boxplot

Barplots of each columns. We use `whis=100.0` to remove whiskers.
`X` can be a list of arrays.

```python
plt.boxplot(X, labels=iris.feature_names, whis = 100.0)
plt.show()
```

## Lines

```python
import random
max_val=10
x_axis=range(0,max_val)
y_axis=[0] * max_val
for i in range(1,max_val):
    y_axis[i]=y_axis[i-1]+random.uniform(-1, 1)
```

```python
plt.plot(x_axis, y_axis);
plt.show()
```
